__Задание 1. (1 балл)__

Для каждого датасета обучите решающее дерево с параметрами по умолчанию, предварительно разбив выборку на обучающую и тестовую. Постройте разделящие поверхности (для этого воспользуйтесь функцией `plot_surface`, пример ниже). Посчитайте accuracy на обучающей и тестовой выборках. Сильно ли деревья переобучились?

````python
def plot_surface(clf, X, y):
    plot_step = 0.01
    palette = sns.color_palette(n_colors=len(np.unique(y)))
    cmap = ListedColormap(palette)
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                         np.arange(y_min, y_max, plot_step))
    plt.tight_layout(h_pad=0.5, w_pad=0.5, pad=2.5)

    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    cs = plt.contourf(xx, yy, Z, cmap=cmap, alpha=0.3)

    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap, alpha=.7,
                edgecolors=np.array(palette)[y], linewidths=2)
````

````python
# Пример:
from sklearn.linear_model import LinearRegression
X, y = datasets[2]
lr  = LinearRegression().fit(X, y)
plot_surface(lr, X, y)
````

````python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score

datasets = [
    make_circles(noise=0.2, factor=0.5, random_state=42),
    make_moons(noise=0.2, random_state=42),
    make_classification(n_classes=3, n_clusters_per_class=1, n_features=2, 
                       class_sep=0.8, random_state=3, n_redundant=0)
]

dataset_names = ['Circles', 'Moons', 'Classification']

plt.figure(figsize=(10, 12))

results = []

for i, ((X, y), name) in enumerate(zip(datasets, dataset_names)):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    
    tree = DecisionTreeClassifier(random_state=42)
    tree.fit(X_train, y_train)
    
    y_train_pred = tree.predict(X_train)
    y_test_pred = tree.predict(X_test)
    
    train_accuracy = accuracy_score(y_train, y_train_pred)
    test_accuracy = accuracy_score(y_test, y_test_pred)
    
    results.append({
        'Dataset': name,
        'Train Accuracy': train_accuracy,
        'Test Accuracy': test_accuracy,
        'Overfitting': train_accuracy - test_accuracy
    })
    plt.subplot(3, 2, i*2 + 1)
    plot_surface(tree, X_train, y_train)
    plt.title(f'{name} - Train (Acc: {train_accuracy:.3f})')
    
    plt.subplot(3, 2, i*2 + 2)
    plot_surface(tree, X_test, y_test)
    plt.title(f'{name} - Test (Acc: {test_accuracy:.3f})')

plt.tight_layout()
plt.show()

for result in results:
    overfitting = result['Overfitting']
    if overfitting > 0.1:
        status = "Сильное переобучение"
    elif overfitting > 0.05:
        status = "Умеренное переобучение"
    else:
        status = "Минимальное переобучение"
    
    print(f"{result['Dataset']}: {status} (разница: {overfitting:.3f})")
````

__Ответ:__все деревья сильно переобучились

__Задание 2. (1.5 балла)__

Попробуйте перебрать несколько параметров для регуляризации (напр. `max_depth`, `min_samples_leaf`). Для каждого набора гиперпараметров постройте разделяющую поверхность, выведите обучающую и тестовую ошибки. Можно делать кросс-валидацию или просто разбиение на трейн и тест, главное делайте каждый раз одинаковое разбиение, чтобы можно было корректно сравнивать (помните же, что итоговое дерево сильно зависит от небольшого изменения обучающей выборки?). Проследите как меняется разделяющая поверхность и обобщающая способность. Почему так происходит, одинаково ли изменение для разных датасетов?

````python
datasets = [
    make_circles(noise=0.2, factor=0.5, random_state=42),
    make_moons(noise=0.2, random_state=42),
    make_classification(n_classes=3, n_clusters_per_class=1, n_features=2, 
                       class_sep=0.8, random_state=3, n_redundant=0)
]

dataset_names = ['Circles', 'Moons', 'Classification']

max_depths = [2, 3, 5, 10, None]  # None - без ограничения глубины
min_samples_leafs = [1, 5, 10, 20]
min_samples_splits = [2, 5, 10]

X_train_all, X_test_all, y_train_all, y_test_all = [], [], [], []

for X, y in datasets:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    X_train_all.append(X_train)
    X_test_all.append(X_test)
    y_train_all.append(y_train)
    y_test_all.append(y_test)

print("Влияние max_depth")
plt.figure(figsize=(20, 15))

results_depth = []

for dataset_idx, (name, (X, y)) in enumerate(zip(dataset_names, datasets)):
    X_train, X_test, y_train, y_test = X_train_all[dataset_idx], X_test_all[dataset_idx], y_train_all[dataset_idx], y_test_all[dataset_idx]
    
    for i, max_depth in enumerate(max_depths):
        tree = DecisionTreeClassifier(max_depth=max_depth, random_state=42)
        tree.fit(X_train, y_train)
        
        train_acc = accuracy_score(y_train, tree.predict(X_train))
        test_acc = accuracy_score(y_test, tree.predict(X_test))
        
        results_depth.append({
            'Dataset': name,
            'max_depth': max_depth,
            'Train_Accuracy': train_acc,
            'Test_Accuracy': test_acc,
            'Difference': train_acc - test_acc
        })
        
        plt.subplot(3, 5, dataset_idx * 5 + i + 1)
        plot_surface(tree, X_train, y_train)

plt.tight_layout()
plt.show()

print("Влияние min_samples_leaf")
plt.figure(figsize=(20, 15))

results_leaf = []

for dataset_idx, (name, (X, y)) in enumerate(zip(dataset_names, datasets)):
    X_train, X_test, y_train, y_test = X_train_all[dataset_idx], X_test_all[dataset_idx], y_train_all[dataset_idx], y_test_all[dataset_idx]
    
    for i, min_samples_leaf in enumerate(min_samples_leafs):
        tree = DecisionTreeClassifier(min_samples_leaf=min_samples_leaf, random_state=42)
        tree.fit(X_train, y_train)
        
        train_acc = accuracy_score(y_train, tree.predict(X_train))
        test_acc = accuracy_score(y_test, tree.predict(X_test))
        
        results_leaf.append({
            'Dataset': name,
            'min_samples_leaf': min_samples_leaf,
            'Train_Accuracy': train_acc,
            'Test_Accuracy': test_acc,
            'Difference': train_acc - test_acc
        })
        
        plt.subplot(3, 4, dataset_idx * 4 + i + 1)
        plot_surface(tree, X_train, y_train)

plt.tight_layout()
plt.show()

print("Оптимальные комбинации параметров")

optimal_params = [
    {'max_depth': 5, 'min_samples_leaf': 5},    # Для Circles
    {'max_depth': 3, 'min_samples_leaf': 10},   # Для Moons  
    {'max_depth': 3, 'min_samples_leaf': 5}     # Для Classification
]

plt.figure(figsize=(15, 5))

for dataset_idx, (name, (X, y)) in enumerate(zip(dataset_names, datasets)):
    X_train, X_test, y_train, y_test = X_train_all[dataset_idx], X_test_all[dataset_idx], y_train_all[dataset_idx], y_test_all[dataset_idx]
    
    tree_optimal = DecisionTreeClassifier(**optimal_params[dataset_idx], random_state=42)
    tree_optimal.fit(X_train, y_train)
    train_acc_optimal = accuracy_score(y_train, tree_optimal.predict(X_train))
    test_acc_optimal = accuracy_score(y_test, tree_optimal.predict(X_test))
    
    plt.subplot(1, 3, dataset_idx + 1)
    plot_surface(tree_optimal, X_train, y_train)

plt.tight_layout()
plt.show()
````

__Ответ:__Регуляризация эффективно борется с переобучением, но оптимальные параметры сильно зависят от структуры данных. Для каждого датасета необходимо подбирать индивидуальные настройки

Протестируйте свое решающее дерево на датасете [mushrooms](https://archive.ics.uci.edu/ml/datasets/Mushroom). Вам нужно скачать таблицу agaricus-lepiota.data (лежит на гитхабе вместе с заданием), прочитать ее с помощью pandas, применить к каждому столбцу LabelEncoder (из sklearn), чтобы преобразовать строковые имена категорий в натуральные числа. Первый столбец — это целевая переменная (e — edible, p — poisonous) Мы будем измерять качество с помощью accuracy, так что нам не очень важно, что будет классом 1, а что — классом 0. Обучите решающее дерево на половине случайно выбранных объектов (признаки в датасете категориальные) и сделайте предсказания для оставшейся половины. Вычислите accuracy.

У вас должно получиться значение accuracy, равное единице (или очень близкое к единице), и не очень глубокое дерево.

````python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from hw2code import *

def load_mushrooms_data():
    
    mushrooms = pd.read_csv('datasets/agaricus-lepiota.data', header=None)
    print(mushrooms.head())
    
    return mushrooms


def test_mushrooms_dataset():
    mushrooms = load_mushrooms_data()
    
    y = mushrooms.iloc[:, 0].values
    X = mushrooms.iloc[:, 1:].values
    
    label_encoders = {}
    X_encoded = np.zeros_like(X, dtype=int)
    
    for i in range(X.shape[1]):
        le = LabelEncoder()
        X_encoded[:, i] = le.fit_transform(X[:, i])
        label_encoders[i] = le
    
    le_target = LabelEncoder()
    y_encoded = le_target.fit_transform(y)
    
    X_train, X_test, y_train, y_test = train_test_split(
        X_encoded, y_encoded, test_size=0.5, random_state=42, stratify=y_encoded
    )
    
    feature_types = ['categorical'] * X_train.shape[1]
    
    our_tree = DecisionTree(feature_types=feature_types, max_depth=5)
    our_tree.fit(X_train, y_train)
    
    y_pred_our = our_tree.predict(X_test)
    accuracy_our = accuracy_score(y_test, y_pred_our)
    
    print(f"Accuracy: {accuracy_our:.4f}")
    print(f"Правильно классифицировано: {np.sum(y_test == y_pred_our)}/{len(y_test)}")

test_mushrooms_dataset()
````

````python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier as SklearnTree
from sklearn.base import BaseEstimator, ClassifierMixin
import matplotlib.pyplot as plt

class MyTreeWrapper(BaseEstimator, ClassifierMixin):
    def __init__(self, feature_types, max_depth=None, min_samples_split=None, min_samples_leaf=None):
        self.feature_types = feature_types
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.min_samples_leaf = min_samples_leaf
        self.tree = None

    def fit(self, X, y):
        self.tree = DecisionTree(
            feature_types=self.feature_types,
            max_depth=self.max_depth,
            min_samples_split=self.min_samples_split,
            min_samples_leaf=self.min_samples_leaf
        )
        self.tree.fit(X, y)
        return self

    def predict(self, X):
        return self.tree.predict(X)

data_url = 'datasets/tic-tac-toe-endgame.csv'
data_frame = pd.read_csv(data_url, header=None)

encoder = LabelEncoder()
target = encoder.fit_transform(data_frame.iloc[:, -1])
features = data_frame.iloc[:, :-1].apply(LabelEncoder().fit_transform).values

feature_categories = ['categorical'] * features.shape[1]

parameter_ranges = {
    'max_depth': range(1, 25),
    'min_samples_split': range(2, 40, 2),
    'min_samples_leaf': range(1, 30, 2)
}

plt.figure(figsize=(18, 5))

accuracy_scores = []
for depth_value in parameter_ranges['max_depth']:
    tree_model = MyTreeWrapper(feature_types=feature_categories, max_depth=depth_value)
    cv_score = cross_val_score(tree_model, features, target, cv=5, scoring='accuracy').mean()
    accuracy_scores.append(cv_score)

plt.subplot(1, 3, 1)
plt.plot(parameter_ranges['max_depth'], accuracy_scores, marker='o')
plt.xlabel('max_depth')
plt.ylabel('Accuracy')
plt.title('max_depth Impact')
plt.grid(True)

accuracy_scores = []
fixed_depth = 10
for split_value in parameter_ranges['min_samples_split']:
    tree_model = MyTreeWrapper(feature_types=feature_categories,
                              max_depth=fixed_depth,
                              min_samples_split=split_value)
    cv_score = cross_val_score(tree_model, features, target, cv=5, scoring='accuracy').mean()
    accuracy_scores.append(cv_score)

plt.subplot(1, 3, 2)
plt.plot(parameter_ranges['min_samples_split'], accuracy_scores, marker='o', color='orange')
plt.xlabel('min_samples_split')
plt.title(f'min_samples_split Impact\n(max_depth={fixed_depth})')
plt.grid(True)

accuracy_scores = []
for leaf_value in parameter_ranges['min_samples_leaf']:
    tree_model = MyTreeWrapper(feature_types=feature_categories,
                              max_depth=fixed_depth,
                              min_samples_leaf=leaf_value)
    cv_score = cross_val_score(tree_model, features, target, cv=5, scoring='accuracy').mean()
    accuracy_scores.append(cv_score)

plt.subplot(1, 3, 3)
plt.plot(parameter_ranges['min_samples_leaf'], accuracy_scores, marker='o', color='green')
plt.xlabel('min_samples_leaf')
plt.title(f'min_samples_leaf Impact\n(max_depth={fixed_depth})')
plt.grid(True)

plt.tight_layout()
plt.show()

def get_dataset(dataset_name):
    label_encoder = LabelEncoder()

    dataset_urls = {
        'mushrooms': 'https://archive.ics.uci.edu/ml/machine-learning-databases/mushroom/agaricus-lepiota.data',
        'tic-tac-toe': 'datasets/tic-tac-toe-endgame.csv',
        'cars': 'https://archive.ics.uci.edu/ml/machine-learning-databases/car/car.data',
        'nursery': 'https://archive.ics.uci.edu/ml/machine-learning-databases/nursery/nursery.data'
    }

    url = dataset_urls[dataset_name]

    if dataset_name == 'mushrooms':
        df = pd.read_csv(url, header=None)
        y = label_encoder.fit_transform(df.iloc[:, 0])
        X_data = df.iloc[:, 1:]

    elif dataset_name == 'tic-tac-toe':
        df = pd.read_csv(url, header=None)
        y = label_encoder.fit_transform(df.iloc[:, -1])
        X_data = df.iloc[:, :-1]

    elif dataset_name == 'cars':
        df = pd.read_csv(url, header=None)
        target_mapping = {'unacc': 0, 'acc': 0, 'good': 1, 'vgood': 1}
        y = df.iloc[:, -1].astype(str).str.strip().map(target_mapping).values
        X_data = df.iloc[:, :-1]

    elif dataset_name == 'nursery':
        df = pd.read_csv(url, header=None)
        target_mapping = {'not_recom': 0, 'recommend': 0, 'very_recom': 1,
                         'priority': 1, 'spec_prior': 1}
        y = df.iloc[:, -1].astype(str).str.strip().map(target_mapping).values
        X_data = df.iloc[:, :-1]

    X_label_encoded = X_data.apply(LabelEncoder().fit_transform).values

    onehot_encoder = OneHotEncoder(sparse_output=False)
    X_onehot = onehot_encoder.fit_transform(X_data)

    return X_label_encoded, X_onehot, y

dataset_names = ['mushrooms', 'tic-tac-toe', 'cars', 'nursery']
comparison_results = {}

for dataset in dataset_names:
    print(f"Working on {dataset}...")
    try:
        X_le, X_ohe, y = get_dataset(dataset)

        performance_scores = []

        real_types = ['real'] * X_le.shape[1]
        dt_real = MyTreeWrapper(feature_types=real_types)
        score_real = cross_val_score(dt_real, X_le, y, cv=10, scoring='accuracy').mean()
        performance_scores.append(score_real)

        cat_types = ['categorical'] * X_le.shape[1]
        dt_cat = MyTreeWrapper(feature_types=cat_types)
        score_cat = cross_val_score(dt_cat, X_le, y, cv=10, scoring='accuracy').mean()
        performance_scores.append(score_cat)

        ohe_types = ['real'] * X_ohe.shape[1]
        dt_ohe = MyTreeWrapper(feature_types=ohe_types)
        score_ohe = cross_val_score(dt_ohe, X_ohe, y, cv=10, scoring='accuracy').mean()
        performance_scores.append(score_ohe)

        sk_model = SklearnTree()
        score_sklearn = cross_val_score(sk_model, X_ohe, y, cv=10, scoring='accuracy').mean()
        performance_scores.append(score_sklearn)

        comparison_results[dataset] = performance_scores

    except Exception as e:
        print(f"Problem with {dataset}: {e}")
        comparison_results[dataset] = [None] * 4

results_dataframe = pd.DataFrame(comparison_results, index=[
    'MyTree Real',
    'MyTree Cat',
    'MyTree OHE',
    'Sklearn'
]).T

results_dataframe
````
